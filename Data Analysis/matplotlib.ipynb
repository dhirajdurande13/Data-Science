```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[11,12,13,14,15]
# creating line graph
plt.plot(x,y)
plt.xlabel('x Label')
plt.ylabel('y Label')
plt.title('Basic Line graph')
plt.show()
```

```python
# creating customized line plot

x=[1,2,3,4,5]
y=[11,12,13,14,15]


def plot_graph(x,y):
    plt.plot(x,y,color='red',linestyle='--',marker="o",linewidth='1',markersize='2')
    plt.xlabel('x Label')
    plt.ylabel('y Label')
    plt.title('Basic Line graph')
    plt.grid() #showing grids in graph
    plt.show()
    
plot_graph(x,y)
```

```python
# creating mutliple plots
x=[1,2,3,4,5]
y=[11,12,13,14,15]
z=[7,8,9,4,3]

# Now we are creating multiple graphs 
plt.figure(figsize=(9,5))

plt.subplot(2,2,1) #first parameter is how many rows of graphs second parameter is how many columns of graphs third parameter include on which index means index of that graph
plt.plot(x,y,color='red')
plt.title('Plot 1')
# plt.show()

plt.subplot(2,2,2)
plt.plot(x,z,color='red')
plt.title('Plot 2')

plt.subplot(2,2,3)
plt.plot(y,z,color='yellow')
plt.title('Plot 3')

plt.subplot(2,2,4)
plt.plot(y,x,color='yellow')
plt.title('Plot 4')
plt.show()
```

```python
# Bar Plot

categories=['A','B','C','D','E']
values=[5,3,6,7,8]
plt.figure(figsize=(5,3))
plt.bar(categories,values,color='orange')
```

```python
# creating histogram
data=[1,2,2,2,2,3,3,4,4,5,5,5]
# bins means dividing 1 to 5 in 5 parts and then show graph 
plt.hist(data,bins=5)
```

```python
# scatter plot
x=[1,2,3,4,5]
y=[4,3,5,6,7]
plt.scatter(x,y,marker="*",color="Orange")
```

```python
value=[40,30,50,10,25]
labels=['A','B','C','D','E']

def plotGraph(value,label):
    explode=[0.2,0,0,0.1,0] #it will move the first slide 
    plt.pie(value,labels=label,explode=explode,autopct="%1.1f%%",colors=['lightgreen','orange','gray','pink','yellow'])
    
    plt.show()
plotGraph(value,labels)
```
