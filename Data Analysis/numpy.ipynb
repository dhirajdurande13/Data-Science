```python
df=pd.read_csv('sales_data.csv')
df.head()
df.columns
```

```python
df.isnull().any()
```

```python
df['Sales_Fill_NA']=df['Total Revenue'].fillna(df['Total Revenue'].mean())

# Werever there is Nan value there it will replace with mean
```

```python
df.head()
```

```python
df=df.rename(columns={'Total Revenue':'Revenue'})
df
```

```python
df.head(10)
```

```python
# Data Aggregation and Grouping
# Finding the mean of the Unit Price according to the region

grouped_mean=df.groupby('Region')['Unit Price'].mean()
print(grouped_mean) 
```

```python
region_group=df.groupby(['Region','Product Name'])['Revenue'].sum()
print(region_group)
```

```python
# Aggregating multiple functions
mult_funct=df.groupby('Region')['Revenue'].agg(['sum','count','mean'])
mult_funct
```
