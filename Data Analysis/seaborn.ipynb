```python
import seaborn as sns
```

```python
# we will load the dataset
tips=sns.load_dataset('tips')
```

```python
tips.head(10)
```

```python
# plotting scatter ploat
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],data=tips)
```

```python
# lineploat
import matplotlib.pyplot as plt
sns.lineplot(x=tips['size'],y=tips['total_bill'],data=tips)
plt.title("Line ploat size vs total bill")
```

```python
tips.head()
```

```python
# barploat
sns.barplot(x=tips['day'],y=tips['total_bill'],data=tips)
plt.title("Bar graph day vs total_bill")
```

```python
# BoxPloat

sns.boxplot(x=tips['day'],y=tips['total_bill'],data=tips)

# out of the box are the outliers and reading start from below 0% 25% 50% 75% 100%
```

```python
# violinploat
# How the distribution of data is done daywise 
sns.violinplot(x=tips['day'],y=tips['total_bill'],data=tips)
```

```python
sns.histplot(tips['total_bill'],bins=5,kde=True)
# Kde is kernal density estimation for showing the curve in between them
```

```python
# if we want to print only kde
sns.kdeplot(tips['total_bill'],fill=True)
# For printing the kernal distribuition density
```

```python
sns.pairplot(data=tips)
# It will show the relationships between diffrent pairs 
# For this specifically needs for numerical data only
```

```python
tips.head()
```

```python
# Heatmap
# finding the realation between numerical data
corr=tips[['total_bill','tip','size']].corr()
```

```python
corr
# Example if total_bill increses by 1 then size will increse by 0.59
```

```python
sns.heatmap(corr,annot=True,cmap="coolwarm")
```
