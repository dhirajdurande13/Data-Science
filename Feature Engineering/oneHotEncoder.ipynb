```python
from sklearn.preprocessing import OneHotEncoder
color=['red','green','blue','yellow','red']
```

```python
import pandas as pd
df=pd.DataFrame(color,columns=['color'])
```

```python
df
```

```python
encoder=OneHotEncoder()
encoded=encoder.fit_transform(df).toarray()
```

```python
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
pd.concat([df,encoder_df],axis=1)
```

```python
# target guided ordinal value

import pandas as pd

# Sample data
data = {
    'city': ['Mumbai', 'Delhi', 'Bangalore', 'Chennai', 'Kolkata','Mumbai'],
    'price': [25000, 22000, 27000, 21000, 20000,3000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display
print(df)
```

```python
df
```

```python
mean_price=df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
df['city_encoded']=df['city'].map(mean_price)
```
